```python
from sympy import sin, cos, tan, atan2, sqrt, Matrix, var, init_printing, simplify, pprint

init_printing(use_unicode=True)

var('tanLambda,d0,phi0,z0', real = True)
var('ptinv', real= True, positive = True)
var('x,y,z,p_x,p_y,p_z', real = True)
var('alpha', real = True)

_ptinv = 1 / sqrt(p_x**2 + p_y**2)  
_tanLambda = p_z / sqrt(p_x**2 + p_y**2) 
_phi0 = atan2(p_y - x * alpha, p_x + y * alpha) 
_z0 = z - p_z * atan2(p_x * x * alpha + p_y * y * alpha, p_x**2 + p_y**2 + p_x * y * alpha - p_y * x * alpha) / alpha 
_area = 2 * (- x * p_y + y * p_x) + alpha * x**2 + alpha * y**2 
_d0 = _area / (sqrt(p_x**2 + p_y**2) + sqrt(p_x**2 + p_y**2 + alpha * _area))


X = Matrix([
    _ptinv, _tanLambda, _phi0, _z0, _d0
])

Y = Matrix([x, y, z, p_x, p_y, p_z])
J = X.jacobian(Y)
J.simplify()

A = J[:, :3]
B = J[:, 3:]
```

```python
var('T,V', real = True, positive = True)
_T = p_x**2 + p_y**2
_P = p_x**2 + p_y**2 + alpha * p_x * y - alpha * p_y *x
_V1 = 2 * (- x * p_y + y * p_x) + alpha * x**2 + alpha * y**2 
_V2 = 2 * alpha * p_x * y - 2 * alpha * p_y * x + alpha**2 * x**2 + alpha**2 * y**2 
```

```python
A = A.subs([(_P, P),(_V1, V),(_V2, alpha * V), (_T, T)])
A
```

```python
B = B.subs([(_P, P),(_V1, V),(_V2, alpha * V), (_T, T)])
B
```

```python
X = X.subs([(_P, P),(_V1, V),(_V2, alpha * V), (_T, T)])
X
```

```python
c0 = X - A * Matrix([x,y,z]) - B * Matrix([p_x, p_y, p_z])
c0
```
